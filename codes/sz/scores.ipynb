```python
import pandas as pd
```

```python
df = pd.read_csv("weight_all.csv")#, encoding='latin-1') 
```

```python
df.head()
```

```python
data = {}
for i in range(1,230):#range(1,230)
    df_1 = df[(df['ID'] == i)]
    df_1c = df_1['纠正后权重']
    sum1 = 0
    for j in list(df_1c):
        try:
            i_f = float(j)
        except ValueError:
            i_f = float(j[:-2])
        sum1 += float(i_f)
    #print(sum1)
    data[i] = sum1
```
